```python
import aula6

menu = "\n--x--\nESCOLHA UMA OPÇÃO:\n1 - Carrega o dataset para uma lista\n2 - Tabela com o título da obra, parte da sua descrição, compositor e ano de criação\n3 - Lista de tuplos (Título, ano) ordenada alfabeticamente\n4 - Lista de tuplos (título, ano) ordenada por ordem crescente\n5 - Distribuição das obras por período\n6 - Distribuição das obras por ano\n7 - Distribuição das obras por compositor\n8 - Gráficos das distribuições\n9 - Lista de compositores associada à sua respetiva lista de obras\n10 - Obras catalogadas no dataset\n0 - Sair\n"

n = 8
while n != 0:
    print(menu)
    n = int(input("Qual a opção do menu?"))
    obras = aula6.lerobras()
    print("A opção escolhida foi:", n)
    if n == 1:
        print(f"{' Lista da informação do dataset ':=^126}\n")
        print(aula6.lerobras())
    if n == 2:
        print(f"{'Tabela das obras':-^81}")
        print(aula6.imp(obras))
    if n == 3:
        print(f"{' Lista de tuplos (Título, ano) ordenada por ordem alfabética ':=^126}")
        print(aula6.ordenaralfa(obras))
    if n == 4:
        print(f"{' Lista de tuplos (Título, ano) ordenada por ordem crescente ':=^126}")
        print(aula6.ordenarc(obras))
    if n == 5:
        print(f"{' Distribuição das obras por período ':=^126}")
        print(aula6.distPeriodo(obras))
    if n == 6:
        print(f"{' Distribuição das obras por compositor ':=^126}")
        print(aula6.disObrasComp(obras))
    if n == 7:
        print(f"{' Distribuição das obras por ano de criação ':=^126}")
        print(aula6.distituloAno(obras))
    if n == 8:
        g = 1
        while g != 0:
            print("1 - Distribuição das obras por período\n2 - Distribuição das obras por compositor\n3 - Distribuição das obras por ano de criação\n0 - Sair da Opção\n")
            g = int(input("Qual a distribuição?"))
            if g == 1:
                print("Distribuição das obras por período")
                obras10 = aula6.distPeriodo(obras)
                print(aula6.distribuicaoP(obras10))
            if g == 2:
                print("Distribuição das obras por compositor")
                obras10 = aula6.compos(obras)
                print(aula6.distribuicaoP(obras10))
            if g == 3:
                print("Distribuição das obras por ano de criação")
                obras10 = aula6.distAno(obras)
                print(aula6.distribuicaoP(obras10))
        if g == 0:
            print("SAIU DA OPÇÃO")
    if n == 9:
        print(f"{' Lista de compositores associada à sua respetiva lista de obras ':=^126}")
        obras11 = aula6.disObrasComp(obras)
        print(aula6.listaObrasComp(obras11))
    if n == 10:
        print(f"{' Obras catalogadas no dataset ':=^126}")
        print(aula6.count())
if n == 0:
    print("O PROGRAMA FECHOU")
```
